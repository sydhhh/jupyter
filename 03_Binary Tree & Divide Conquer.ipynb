        = 0 + logn\*O(n) = O(nlogn)

    - 正好二叉树就是这样的，所以复杂度为O(n)

## 遍历模板（一个人在战斗）

- 分成两个函数写：
    - __俩函数外面定义全局变量result__
    - __第一个函数里面调用一次第二个函数，然后下一行就return出结果了（不用改，就这两行背一下就好了）__
    - __第二个函数是void，递归调用自己，实现算法__
        - 第一行当然是递归出口，直接return，无返回值。
        
```python
class Solution:
    result = [] #Traversal要有一个全局变量
    
    def preorderTraversal(self, root): #这个函数就是个空壳，但必须要有它来引入root，并且return最终结果result
        self.traverse(root)
        return self.result

    def traverse(self, root):
        if root is None: #表示traverse到头了，return空即可
            return
        self.result.append(root.val) #这是preorder，所以这一句在最前面。inorder就在中间，postorder在最后
        self.traverse(root.left)
        self.traverse(root.right)




# 下面是另外一种写法，让result当递归的输入，相当于把非全局变量作为全局变量了（要能看懂!）        
class Solution:
    def preorderTraversal(self, root): 
        result = []
        self.traverse(root, result)
        return result
        
    def traverse(self, root, result):
        if root is None:
            return
        result.append(root.val) 
        self.traverse(root.left, result)
        self.traverse(root.right, result)
        
        

```

## 分治算法模板（找俩小弟做）

- <font color='red'>80%的题可以用这个模板解决!</font>
- 简单来说就是把事情给小弟做，然后老大统计小弟的结果：
    - 把一个复杂的问题分成两个或更多的相同或相似的子问题，再把子问题分成更小的子问题……直到最后子问题可以简单的直接求解，原问题的解即子问题的解的__合并__。
    > 前序遍历，根让左右俩小弟去做它们的前序遍历：
        - 左小弟说：我这边前序遍历是2-4-5
        - 右小弟说：我这边前序遍历是3
        - 根说：那么我们汇总起来吧，我是1，所以最后1-2-4-5-3





- __包含四步：__（注意九章答案里面的注释可能有问题，把conquer写到divide里了）
    - 可以不用递归直接解决的base case，也就是__递归的出口__: (一般就是root为空，有时候还要单独考虑叶子节点)
    - Divide: 拆解（二叉树问题天然拆解好了，所以不用写）
    - Conquer: 递归
    - Merge: 汇总起来，然后最终结果在return里面
    
```python
class Solution:
    def preorderTraversal(self, root):
        result = []
    # 当前为空节点时，说明递归到头了，return当前的空result
        if root is None:
            return result 
    #Divide：二叉树天生给你divide好了，所以这里为空  
    #Conquer
        left = self.preorderTraversal(root.left) #注意，left左边没有self. 
        right = self.preorderTraversal(root.right)
    #Merge 
        result.append(root.val) #改一下这行的顺序就可以得到其他两种
        result.extend(left) #注意是extend而不是append，这里是加一整个list而不是一个元素
        result.extend(right)
        return result
        
```

    - 非递归一般不用

### 遍历法和分治法区别：
- 而遍历法是自己解决问题
    - Top down
    - __本质是有全局变量: 一般就叫它Result __
    - __递归的出口return空__
- 分治法感觉是无脑把问题丢给左右俩子树
    - Bottom up
    - __本质是无全局变量，result在每次递归第一行新建__
    - __递归的出口就return当前的结果__
        - 递归不下去时return（空）result
        - 最后一行还要return result
       

### <font color='red'>遍历法模板记忆方法（只在空和叶子时return void）</font>
```python
#-----------------------------------------
def traverse(node): #void
    if node is None:
        return
    if node is leaf:
        # Do something with global paramater
        return
    
    # 下面开始就是表明node既不是None也不是leaf
    # Do something 
    traverse(node.left)
    # Do something 
    traverse(node.right)
    # Do something 
#-----------------------------------------

```
- 1.`def`外初始化result最为全局变量
- 2.第一个`def`就两行：
    -  调用第二个`def`
    -  `return self.result`
- 3.第二个`def`递归调用自己：
    - 通常输入时还要多一个current变量，见例题1的curDepth，例题2的curPath：
        - 下一层是当前`Depth++`
        - 下一层是当前`Path += "root.left.val"`
    - 递归的出口：一般就考虑root为空时，写作`return` 
        - 不是`return result`
        - 有时还得考虑其他情况，如叶子节点，还是写作`return`，因为最终结果在全局变量里面。
    - 实现算法，考虑整棵树在该问题上的结果，和左右子树之间的联系，对`self.result`进行修改

### <font color='red'>分治法模板记忆方法（空，叶子和最后都return result）</font>

```python
#-----------------------------------------
def traverse(node): # not void
    result = [] #不一定要做
    if node is None:
        return result
    if node is leaf:
        # Do something 
        return result
    
    # 下面开始就是表明node既不是None也不是leaf
    #Conquer：left和right都是当前node得到的最终结果，因此更容易理解
    left = traverse(node.left)
    right = traverse(node.right)
    # Merge：左子树结果，右子树结果，和根节点怎么构成上一层问题的结果
    result = Merge from left and right
    return result
#-----------------------------------------
    ```
- 三个return怎么理解？
    - 每次递归时只可能return一次
    - 要么当前为空node
    - 要么叶子要特别考虑
    - 最后那个return是处理非空和非叶子的情况
    
    
- 步骤
    - 1.初始化result (不一定要做)
    - 2.递归的出口：即可以不递归解决的base case
        - 一般就考虑root为空时，写作`return result`，其实就是return一个空结果
        - 有时还得考虑其他情况，如叶子节点，还是写作`return result`，当然这个时候result非空
        - 有时甚至可能`return`，比如我们不需要result，见例6。
    - 3.Conquer：（Divide默认做好了）就是无脑扔给左右俩子树去做递归
    - 4.Merge：把答案汇总，最后还是写作`return result`。
        - 这是真正实现算法的步骤，考虑整棵树在该问题上的结果，和左右子树之间的联系
    

### <font color = 'red'>分治法遍历法结合模板</font>
- 首先还是遍历法的模板
- 在遍历法模板的第二个`def`里面用分治的思想 
- <font color='red'>核心在于分治法得到的不是全局的结果，但是在递归过程中每一步和全局变量有比较，这样就可以把全局结果记录下来了。</font>
    - 比如例3里面，找全局最小的`subtree`之和：而分治法`subsum(root)`得到的是当前节点下的`sum`之和，必须有一个别的变量来记录不同的`node`在调用`subsum(node)`之后的结果并找到最小值。


### <font color='red'>ResultType</font>
- 有的时候还需要return多个值：
    - java: `class ResultType { int var1, var2; }`
    - 其实python没这个问题，因为return可以return两个值，java只能新写个类，才能return两个值。
- 见例题4，5

# 二叉搜索树 Binary Search Tree

- 左子树都比根节点小，右子树都大于等于
- in-order traversal是“__不下降__”序列
- 功能：
    - Insert / Remove / Find / Validate
    - <font color='red'>本质上就是上一个Note里面的Binary Search</font>
        - 先比较`root.val`和`target`，然后决定去左边还是右边，这样进行下去就可以找到`target`了。

# 上课例题
## `Maximum Depth of Binary Tree (分治好)` 
http://www.lintcode.com/problem/maximum-depth-of-binary-tree/

http://www.jiuzhang.com/solutions/maximum-depth-of-binary-tree/

- 先考虑整棵树在该问题上的结果，和左右子树在该问题上的结果之间的联系。
    - 然后不管三七二十一，先用分治法试试。


```python
# 分治法，极度简洁
# 总问题的结果不就是左右子树的最大深度中的大的那个，加上1么！
class Solution:
    def maxDepth(self, root):
        if root is None: #Exit the recursion
            return 0
        
        leftMaxDepth = self.maxDepth(root.left)
        rightMaxDepth = self.maxDepth(root.right)
        
        return max(leftMaxDepth, rightMaxDepth) + 1 #Merge
        
# ------------------------------------------------------------        
# 遍历法：不好想
# 全局变量就是最终答案result
# 递归时要多一个curDepth作为输入。
class Solution:
    result = 0
    def maxDepth(self, root):
        self.traverse(root, 1)
        return self.result
        
    def traverse(self, root, curDepth):
        if root is None:
            return 
        if curDepth > self.result:
            self.result = curDepth
        self.traverse(root.left, curDepth+1)
        self.traverse(root.right, curDepth+1)
```


## `Binary Tree Paths (遍历好)`
http://www.lintcode.com/en/problem/binary-tree-paths/

http://www.jiuzhang.com/solutions/binary-tree-paths/

>  return all root-to-leaf paths.

```python

# 分治法
class Solution:
    def binaryTreePaths(self, root):
        #每次递归第一行新建一个空的结果。
        Paths = [] 
        
        # 递归的出口：即可以不递归解决的base case
            # 一般只要考虑None即可
            # 这里还要单独考虑叶子节点，没有"->"
        if root is None: #递归不下去了，return当前的空Paths
            return Paths
        if root.left is None and root.right is None:
            Paths.append(str(root.val))
            return Paths
         
        # Conquer: 丢给左右子树
        leftPaths = self.binaryTreePaths(root.left) 
        rightPaths = self.binaryTreePaths(root.right)
        
        #Merge: 注意是for循环，因为leftPaths里面有很多Paths，然后每一条都要在开头加上当前的root
        for Path in leftPaths:
            Paths.append(str(root.val) + "->" + Path)
        for Path in rightPaths:
            Paths.append(str(root.val) + "->" + Path)
        return Paths

#----------------------------------------------------
#遍历法：这时候遍历法的思路更清晰，没有for循环了：
    # 分治时左子树得到一堆path，然后要把root一一加到这些path上
    # 遍历时因为就是一步步找下去，很直接不需要for
class Solution:
    # 初始化全局变量
    Paths = []
    def binaryTreePaths(self, root):
        if root:
            self.traverse(root, str(root.val)) #因为要保证root.val作为curPath存在
        return self.Paths

    def traverse(self, root, curPath):
        # 递归的出口，只return
        # 叶子节点单独考虑
        if root is None:
            return
        if root.left is None and root.right is None:
            self.Paths.append(curPath)
            return
        
        #调用自己，这个if判断语句还是为了保证.val存在
        if root.left:
            self.traverse(root.left, curPath+'->'+str(root.left.val))
        
        if root.right:
            self.traverse(root.right, curPath+'->'+str(root.right.val))    

```

## `Minimum Subtree (遍历分治结合)`
http://www.lintcode.com/en/problem/minimum-subtree/

http://www.jiuzhang.com/solutions/minimum-subtree/

> Given a binary tree, find the subtree with minimum sum. Return the root of the subtree.

- 不用ResultType：
    - 本质还是分治
    - 分治法得到的是subTree的sum值，但我们其实不需要它
    - 需要全局变量来得到最小的那个sum以及对应的Node在哪，这就是遍历的思想
```python
import sys
class Solution:
    resultNode = None
    Minsum = sys.maxint 
    
    def findSubtree(self, root):
        totalSum = self.subsum(root)
        return self.resultNode
        
    def subsum(self, root):
        if root is None:
            return 0
        sum = root.val + self.subsum(root.left) + self.subsum(root.right)
        
        if sum < self.Minsum:
            self.Minsum = sum
            self.resultNode = root
        return sum
        ```

## `Subtree with Maximum Average (ResultType)`
http://www.lintcode.com/problem/subtree-with-maximum-average/

http://www.jiuzhang.com/solutions/subtree-with-maximum-average/

> Given a binary tree, find the subtree with maximum average. Return the root of the subtree.

- 和上一题很像，但因为是平均值，所以要`return sum, size`
```python
import sys
class Solution:
    resultNode = None
    MaxAve = - sys.maxint
    def findSubtree2(self, root):
        totalAve = self.subAve(root)
        return self.resultNode
    
    def subAve(self, root):
        if root is None:
            return 0, 0
        
        leftSum, leftSize = self.subAve(root.left)
        rightSum, rightSize = self.subAve(root.right)
        
        sum, size = leftSum + root.val + rightSum, \
                    leftSize + 1 + rightSize
                    
        if sum > size * self.MaxAve:
            self.MaxAve = sum / size
            self.resultNode = root
        
        return sum, size
        ```

## `Balanced Binary Tree (ResultType)` 
http://www.lintcode.com/problem/balanced-binary-tree/

http://www.jiuzhang.com/solutions/balanced-binary-tree/

> 判断是否balanced

- 从分治的思想上来说，首先要求左子树和右子树都平衡，同时左右子树高度差也不大于1
    - 所以还需要return当前高度
    
```python
class Solution:
    def isBalanced(self, root):
        result,_ = self.validate(root)
        return result
        
    def validate(self, root):
        if root is None:
            return True, 0
        
        leftBalanced, leftHeight = self.validate(root.left)
        if not leftBalanced:
            return False, 0
        
        rightBalanced, rightHeight = self.validate(root.right)
        if not rightBalanced:
            return False, 0
            
        return abs(leftHeight - rightHeight) <= 1, max(leftHeight, rightHeight) + 1 
        ```

## `Flattern Binary Tree to Linked List (DV时return空)`
http://www.lintcode.com/problem/flatten-binary-tree-to-linked-list/

http://www.jiuzhang.com/solutions/flatten-binary-tree-to-linked-list/

- 一般用分治法要求return value，但这时候我们不需要result，我们只是对原来的树进行操作，所以return空！

``` python

class Solution:
    def flatten(self, root):
        # 第一步也不用初始化，因为没有result
        # 递归出口：
        if root == None:
            return 
        #Conquer：另一方面想因为conquer之后不需要赋值给谁，所以return那里就是空了
        self.flatten(root.left)
        self.flatten(root.right)
        #Merge
        p = root
        if p.left is None:
            return 
        p = p.left
        while p.right:
            p = p.right
        p.right = root.right
        root.right = root.left
        root.left = None
```

    - left和right都非空，那就说明A和B在当前根的左右子树上，return root
    - 最后一种情况就是找不到A也找不到B，更不用说LCA了，return None。

```python
class Solution:
    def lowestCommonAncestor(self, root, A, B):
        #没必要初始化
        #递归出口，安装上面的分析，需要考虑空或者正好撞上
        if root is None or root is A or root is B:
            return root
        
        #Conquer
        left = self.lowestCommonAncestor(root.left, A, B)
        right = self.lowestCommonAncestor(root.right, A, B)
        
        # Merge
        if left and right:
            return root
        if left:
            return left
        if right:
            return right
        return None
        ```

### `Lowest Common Ancestor II`
http://www.lintcode.com/en/problem/lowest-common-ancestor-ii/

> 给了你parent指针

- 直接倒推，非常简单，不用DFS

```python
class Solution:
    def lowestCommonAncestorII(self, root, A, B):
        dict = {}
        while A:
            dict[A] = True
            A = A.parent
        
        while B:
            if B in dict:
                return B
            B = B.parent
            
        return root
        ```

### `Lowest Common Ancestor III`
http://www.lintcode.com/en/problem/lowest-common-ancestor-iii/

http://www.jiuzhang.com/solutions/lowest-common-ancestor-iii/

> node A or node B may not exist in tree.

- 那就不能偷懒像I一样把LCA当作A或B是否存在了，这个时候要return多个值。
    - 情况就很复杂了

```python
class Solution:
    def lowestCommonAncestor3(self, root, A, B):
        a, b, lca = self.helper(root, A, B)
        if a and b:
            return lca
        else:
            return None
            
    def helper(self, root, A, B):
        if root is None: 
            return False, False, None
        
        left_a, left_b, left_lca = self.helper(root.left, A, B)
        right_a, right_b, right_lca = self.helper(root.right, A, B)
        
        #Merge
        a = left_a or right_a or root == A
        b = left_b or right_b or root == B 
        
        if root == A or root == B:
            return a, b, root
        if left_lca and right_lca :
            return a, b, root
        if left_lca:
            return a, b, left_lca
        if right_lca:
            return a, b, right_lca
        return a, b, None
        ```

### `Binary Tree Longest Consecutive Sequence I`
http://www.lintcode.com/problem/binary-tree-longest-consecutivesequence/

http://www.jiuzhang.com/solutions/binary-tree-longest-consecutive-sequence/

> Given a binary tree, find the length of the longest consecutive sequence path.(数值上连续)

> cannot be the reverse

- 使用遍历分治结合：（和例3很像）

```python
class Solution:
    longest = 0 #全局变量
    def longestConsecutive(self, root):
        self.helper(root) #这里return的值其实不需要，因为这代表以root为根的最长，而不是全局最长
        return self.longest
        
    def helper(self, root):
        #不需要初始化
        #递归出口
        if root is None:
            return 0
        left = self.helper(root.left)
        right = self.helper(root.right)
        
        #Merge
        subtreeLongest = 1 # At least we have root
        if root.left and root.val + 1 == root.left.val:
            subtreeLongest = max(subtreeLongest, left+1)
        if root.right and root.val + 1 == root.right.val:
            subtreeLongest = max(subtreeLongest, right+1)
        
        if(subtreeLongest>self.longest):
            self.longest = subtreeLongest
        
        return subtreeLongest
        ```

### `Binary Tree Longest Consecutive Sequence II`
> 可以拐弯

### `Binary Tree Longest Consecutive Sequence III`
> 可以拐弯且是多叉树

### `Binary Tree Path Sum I`
http://www.lintcode.com/problem/binary-tree-path-sum/

http://www.jiuzhang.com/solutions/binary-tree-path-sum/


> Given a binary tree, find all paths that sum of the nodes in the path equals to a given number target.

> 从根出发，叶子结束

- 用遍历挺直观的，每次遍历到头就生成一条路径，看看和是否满足。
- 这时两个重点：
    - `curPath.pop()`: 
        - 这里不pop的话，curPath永远越来越多，不能回溯了！
        - <font color='red'>注意例2里面不用`pop()`！为啥呢？</font>
            - `self.traverse(root.left, curPath+'->'+str(root.left.val))`并没有改变`curPath`的值
            - 而本例需要`curPath.append(root.left.val)`,所以需要`pop`来回溯
            - 所以如果不想用`pop()`:
 
            ```python
            #curPath.append(root.left.val)
            #self.helper(root.left, curPath, target-root.left.val)
            #curPath.pop()
            tmp = curPath[:]
            tmp.append(root.left.val)
            self.helper(root.left, tmp, target-root.left.val)
            ```
            
    - `self.result.append(curPath[:])`：<font color='red'>必须是deep copy，卡了我大半天！！！</font>


- 具体实现
```python
class Solution:
    result = [] #全局变量
    def binaryTreePathSum(self, root, target):
        if root:
            curPath =  [root.val]
            self.helper(root, curPath, target-root.val)
        return self.result
    
    def helper(self, root, curPath, target):
        if root is None: #递归出口
            return
        if root.left is None and root.right is None and target==0: #满足题目条件
            self.result.append(curPath[:])
        if root.left: #traverse过程
            curPath.append(root.left.val)
            self.helper(root.left, curPath, target-root.left.val)
            curPath.pop() #别忘了pop()   
        if root.right:
            curPath.append(root.right.val)
            self.helper(root.right, curPath, target-root.right.val)
            curPath.pop()
  ```


- 也可以用分治做，这个时候子问题就是在左子树里面找到所有的路径之和为target-root.val的path，这样不可避免的就像例2一样，要用for循环把当前root加入left的所有结果中。

### `Binary Tree Path Sum II`
> 不一定从根开始，叶子结束

- 用遍历简单，分治法特别麻烦（虽然分治法覆盖面更广一点。）

### `Binary Tree Path Sum III`
> 可以拐弯
